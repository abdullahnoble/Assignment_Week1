# Importing libraries and opening files

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('iris.csv')
df.head()
```

**SepalLength, SepalWidth, PetalLength and PetalWidth** are numerical,
**Name** is nominal.

# Function to calculate and display Summary statistics

```python
def display_summary_stats(var):
    s_len = var
    mean = s_len.mean()
    max1 = s_len.max()
    min1 = s_len.min()
    Range = max1 - min1
    SD = s_len.std()
    variance = SD*SD
    med = s_len.median()
    p25 = s_len.quantile(0.25)
    p75 = s_len.quantile(0.75)
    count = len(s_len)
    print("Maximum value: ",max1)
    print("Minimum value: ",min1)
    print("Range: ",Range)
    print("Number of entries: ",count)
    print("Mean: ",mean)
    print("Standard Deviation: ",SD)
    print("Variance: ",variance)
    print("Median: ",med)
    print("25 percentile value: ",p25)
    print("75 percentile value: ",p75)
```

## Summary stats for SepalLength

```python
display_summary_stats(df['SepalLength'])

SLength = df['SepalLength']
x1 = np.arange(0.0,8.5,0.3)
plt.hist(SLength,x1, histtype='bar',rwidth=0.9)
plt.xlabel('Length of Sepal')
plt.ylabel('Number of readings')
plt.title("Histogram for Sepal Length")
plt.show()
```

## Summary stats for SepalWidth

```python
display_summary_stats(df['SepalWidth'])

SWidth = df['SepalWidth']
x2 = np.arange(0.1,8.0,0.3)
plt.hist(SWidth,x2, histtype='bar',rwidth=0.9)
plt.xlabel('Width of Sepal')
plt.ylabel('Number of readings')
plt.title("Histogram for Sepal Width")
plt.show()
```

## Summary stats for PetalLength

```python
display_summary_stats(df['PetalLength'])

PLength = df['PetalLength']
x3 = np.arange(0.0,8.0,0.3)
plt.hist(PLength,x3, histtype='bar',rwidth=0.9)
plt.xlabel('Length of Petal')
plt.ylabel('Number of readings')
plt.title("Histogram for Petal Length")
plt.show()
```

## Summary stats for PetalWidth

```python
display_summary_stats(df['PetalWidth'])

PWidth = df['PetalWidth']
x4 = np.arange(0.1,8.0,0.3)
plt.hist(PWidth,x4, histtype='bar',rwidth=0.9)
plt.xlabel('Width of Petal')
plt.ylabel('Number of readings')
plt.title("Histogram for Petal Width")
plt.show()
```

# Boxplot

```python
plt.boxplot([SLength,SWidth,PLength,PWidth])
plt.ylabel('Range of values')
```
